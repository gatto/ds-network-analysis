# Sentiment Analysis

```python
df_copy=df.set_index('user_id')
```

```python
df_copy=df_copy[~df_copy.index.duplicated(keep='first')]
df_copy.shape
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
#manca pulire tweets
```

```python
scores = []
# Declare variables for scores
compound_list = []
positive_list = []
negative_list = []
neutral_list = []
for i in range(df['text'].shape[0]):
#print(analyser.polarity_scores(sentiments_pd['text'][i]))
    compound = analyzer.polarity_scores(df['text'][i])["compound"]
    pos = analyzer.polarity_scores(df['text'][i])["pos"]
    neu = analyzer.polarity_scores(df['text'][i])["neu"]
    neg = analyzer.polarity_scores(df['text'][i])["neg"]
    
    scores.append({"Compound": compound,
                       "Positive": pos,
                       "Negative": neg,
                       "Neutral": neu
                  })
    
sentiments_score = pd.DataFrame.from_dict(scores)
df = df.join(sentiments_score)
```

```python
df=df.set_index('user_id')
df
```

```python
# create a list of our conditions
conditions = [
    (df['Compound'] <= -0.5),
    (df['Compound'] > -0.5) & (df['Compound'] < 0.5),
    (df['Compound'] > 0.5)
    ]

# create a list of the values we want to assign for each condition
values = ['Negative', 'Neutral', 'Positive']

# create a new column and use np.select to assign values to it using our lists as arguments
df['Category'] = np.select(conditions, values)
df.head()
```

```python
type(df['Category'][290883])
```

```python
users=set(df.index)
for u_set in users:
    user=[]
    cat1=[]
    cat2=[]
    pos=0
    neg=0
    neutr=0
    for u in df.index:
        if u_set==u:
            if [df['Category'][u]=='Positive']==True:
                pos += 1 
            elif [df['Category'][u]=='Negative']==True:  
                neg += 1 
            elif [df['Category'][u]=='Neutral']==True:
                neutr +=1 
        if max(pos,neg,neutr)==pos:
            cat2.append('Positive')
        elif max(pos,neg,neutr)==neg:
            cat2.append('Negative')
        elif max(pos,neg,neutr)==neutr:
            cat2.append('Neutral')       
        user.append(u_set)
        cat1.append(df['category'][u_set])  
```

```python
final_df=pd.DataFrame({'user_id': user,
                      'cat1':cat1,
                      'cat2':cat2})
final_df
```

```python
final_df.to_csv('df_sentiment')
```
