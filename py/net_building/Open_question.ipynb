```python
import socket
from urllib3.connection import HTTPConnection

HTTPConnection.default_socket_options = (
    HTTPConnection.default_socket_options + [
        (socket.SOL_SOCKET, socket.SO_KEEPALIVE, 1),
        (socket.SOL_TCP, socket.TCP_KEEPIDLE, 45),
        (socket.SOL_TCP, socket.TCP_KEEPINTVL, 10),
        (socket.SOL_TCP, socket.TCP_KEEPCNT, 6)
    ])

def trad(stringa):
    translated = GoogleTranslator(source='auto', target='en').translate(stringa)
    return translated
```

```python
df1['text']=df1['text'].map(trad)
```

```python
df1.to_csv('tweets_trad_parte1.csv')
```

```python
df2['text']=df2['text'].map(trad)
```

```python
df2.to_csv('tweets_trad_parte2.csv')
```
