```python
m = SocialDB(100)
```

```python
g= list(m.df.index)
G = nx.Graph()
G.add_nodes_from(g, stand=m.df["class"])
G.nodes[195406]["stand"]
```

```python
edges = [(x["from"], x["to"]) for x in m.edges]
print(edges)
```

```python
G.add_edges_from(edges)
```

```python
G.number_of_edges()
```

```python
G.number_of_nodes()
```

```python
nx.draw(G)
```

```python
degree_sequence = sorted((d for n, d in G.degree()), reverse=True)
dmax = max(degree_sequence)

fig = plt.figure("Degree of a random graph", figsize=(8, 8))
# Create a gridspec for adding subplots of different sizes
axgrid = fig.add_gridspec(5, 4)

ax0 = fig.add_subplot(axgrid[0:3, :])
Gcc_G = G.subgraph(sorted(nx.connected_components(G), key=len, reverse=True)[0])
pos = nx.spring_layout(Gcc_G, seed=10396953)
nx.draw_networkx_nodes(Gcc_G, pos, ax=ax0, node_size=20)
nx.draw_networkx_edges(Gcc_G, pos, ax=ax0, alpha=0.4)
ax0.set_title("Connected components of G")
ax0.set_axis_off()

ax1 = fig.add_subplot(axgrid[3:, :2])
ax1.plot(degree_sequence, "b-", marker="o")
ax1.set_title("Degree Rank Plot")
ax1.set_ylabel("Degree")
ax1.set_xlabel("Rank")

ax2 = fig.add_subplot(axgrid[3:, 2:])
ax2.bar(*np.unique(degree_sequence, return_counts=True))
ax2.set_title("Degree histogram")
ax2.set_xlabel("Degree")
ax2.set_ylabel("# of Nodes")

fig.tight_layout()
plt.show()
```

### Create an Erdons Renyi Random Graph
